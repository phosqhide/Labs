```python
import urllib.request
import pandas as pd
import os
from datetime import datetime
import re
```

## Download files


```python
def download_csv(id):
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    url=f'https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={i}&year1=1981&year2=2024&type=Mean'
    try:
        vhi_url = urllib.request.urlopen(url)
        os.makedirs("data", exist_ok=True)

        filepath = os.path.join("data", f"vhi_id_{id}_{timestamp}.csv")
        with open(filepath, 'wb') as out:
            out.write(vhi_url.read())

        print(f"Id {id} is download in {filepath}")
        
    except Exception as e:
        print(f"Your problem id is {id} and error:{e}")

for i in range(1,28):
    download_csv(i)
```

## Read File and Clean


```python
def clean_html_tags(df):
    for column in df.columns:
        df[column] = df[column].astype(str).str.replace(r'<.*?>', '', regex=True)
    return df

def read_files_to_dataframe(directory_path):
    headers = ['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI', 'Empty']


    all_data = []

    for filename in os.listdir(directory_path):
        if filename.endswith(".csv"):
            filepath = os.path.join(directory_path, filename)
            print(filepath)
            
            df = pd.read_csv(filepath, header=1, names=headers, index_col=False)
            df = clean_html_tags(df)
            df = df.drop(df.loc[df['VHI'] == -1].index)
            df = df.drop("Empty", axis=1)
            df = df.drop(df.index[-1]) 
            
            region_id = filename.split('_')[2]  
            df['area'] = region_id
            
            
            all_data.append(df)

    
    main_df = pd.concat(all_data, ignore_index=True)
    main_df['Week'] = main_df['Week'].astype(float).astype(int)
    main_df['area'] = main_df['area'].astype(int)
    main_df = main_df.sort_values(by=['area', 'Year', 'Week']).reset_index(drop=True)
    return main_df


directory = 'data'  
df = read_files_to_dataframe(directory)
```

## Пошук екстремумів (min та max) для вказаних областей та років, середнього, медіани;

```python
def select_min(index, year):
    return df[(df["area"] == index) & (df["Year"] == year)]['VHI'].min()
    
def select_max(index, year):
    return df[(df["area"] == index) & (df["Year"] == year)]['VHI'].max()
    
def select_average(index, year):
    df["VHI"] = pd.to_numeric(df["VHI"], errors='coerce')

    filtered = df[(df["area"] == index) & (df["Year"] == year)]

    if filtered.empty:
        return 0
        
    return filtered['VHI'].mean()

def select_median(index, year):
    return df[(df["area"] == index) & (df["Year"] == year)]['VHI'].median()
```

## Ряд VHI за вказаний діапазон років для вказаних областей;

```python
def select_VHI_with_years(areas, start_year, end_year):
    filtered = df[(df["area"].isin(areas)) & (df["Year"] >= start_year) & (df["Year"] <= end_year)]
    
    if filtered.empty:
        return None 
    
    return filtered[['area', 'Year', 'VHI']]
```

```python
print(select_VHI_with_years([5,2], start_year=2023, end_year=2024))
```

## Для всього набору даних виявити роки, протягом яких екстремальні посухи торкнулися більше вказаного відсотка областей по Україні (20% областей - 5 областей з 25). Повернути роки, назви областей зекстремальними посухами та значення VHI;

```python
def extreme_drought(percent):
    df_drought = df[(df["VHI"] <= 15) & (df["VHI"] != -1)]
    year_group = df_drought.groupby('Year')['area'].nunique()
    #print(year_group)
    total_areas = df['area'].nunique()
    #print(total_areas)
    percenent = total_areas * percent
    years_extreme_drought = year_group[year_group >= percenent].index
    result = df_drought[df_drought['Year'].isin(years_extreme_drought)][['area', 'Year', 'VHI']]
    
    return result


    
extreme_drought(percent=0.2)
```
